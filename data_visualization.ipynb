```python
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Geo
from pyecharts.faker import Faker
from pyecharts.globals import ChartType
import os
```

```python
df=pd.read_csv('百度球队三元组.csv')
df=df.iloc[20:40]

df['object'][38]="西班牙首都马德里1"
df['object'][39]="西班牙首都马德里2"
df
```

```python
location_dic={
    "西班牙安达鲁西亚自治区加的斯":[-6.18,36.32],
    "西班牙巴伦西亚自治区埃尔切":[-0.42,38.15],
    "西班牙安达卢西亚自治区首府塞维利亚":[-5.59,37.23],
    "西班牙纳瓦拉自治区首府潘普洛纳":[-1.38,42.49],
    "西班牙巴塞罗那市":[2.05,41.23],
    "西班牙安达鲁西亚自治区格拉纳达省省会格拉纳达":[-3.41,37.13],
    "西班牙巴伦西亚自治区卡斯特利翁省比利亚雷亚尔市":[-0.06,39.56],
    "西班牙北部巴斯克自治区比斯开省毕尔巴鄂市":[-2.58,43.15],
    "西班牙瓦伦西亚":[-0.22,39.28],
    "西班牙加利西亚自治区比戈市":[-8.43,42.14],
    "西班牙巴斯克地区的圣塞巴斯提安":[-3.21,41.88],
    "西班牙加泰罗尼亚自治区巴塞罗那市":[2.11,41.23],
    "西班牙安达鲁西亚首府塞维利亚":[-5.59,37.23],
    "西班牙首都马德里":[-3.39,40.24],
    "西班牙首都马德里1":[-3.40,40.20],
    "西班牙首都马德里2":[-3.42,40.30],
    "西班牙马略卡帕尔马":[2.39,39.34],
    "西班牙东部巴伦西亚市":[-0.22,39.28],
    "西班牙马德里自治区赫塔费":[-3.42,40.18],
    "西班牙北部巴斯克自治区维多利亚":[-2.40,42.51]
}
```

```python
#
# 基础数据

keys=df['object']
values=df['subject']

c=Geo()

for (i,j) in location_dic.items():
    c.add_coordinate(i,j[0],j[1])

c.add_schema(maptype="西班牙")
c.add(
        "球队所在地",
        [list(z) for z in zip(keys, values)],
        type_=ChartType.EFFECT_SCATTER,
    )
c.set_series_opts(label_opts=opts.LabelOpts(is_show=False))
c.set_global_opts(title_opts=opts.TitleOpts(title="西甲全部球队所在地"))
c.render()

# 打开html
os.system("render.html")
```
