```python
import re
import requests
from lxml import etree
from urllib.request import urlopen
import urllib.parse
from bs4 import BeautifulSoup
import csv
import bs4
import pandas as pd
import numpy as np
```

```python
url = 'https://zh.wikipedia.org/zh-sg'
r = requests.get('https://zh.wikipedia.org/zh-sg/%E8%A5%BF%E7%8F%AD%E7%89%99%E8%B6%B3%E7%90%83%E7%94%B2%E7%BA%A7%E8%81%94%E8%B5%9B')
```

```python
print(r.text)
```

```python
html = etree.HTML(r.text)
table = html.xpath('//*[@id="mw-content-text"]/div[1]/table[2]/tbody//tr')
print(len(table))
teams = []
for i in range(20):
    tr = table[1+i]
    tname = tr.xpath('string(.)').split()[0]
    td = tr.xpath('.//td[1]//a/@href')[0][5:]
    teams.append((tname, td))
print(teams)
```

```python
teamsBreafInfo = {}
for i in range(len(teams)):
# for i in range(1):
    print(teams[i][0])
    teamurl = url + teams[i][1]
    print(teamurl)
    res = requests.get(teamurl)
    html = etree.HTML(res.text)
    breafinfo = {}
    trnum = html.xpath('//*[@class="infobox vcard"]/tbody//tr').__len__()
    for j in range(2, trnum-4):
        tr = html.xpath('//*[@class="infobox vcard"]/tbody//tr['+str(j)+']')[0]
        th = tr.xpath('.//th[1]//text()')[0]
        td = tr.xpath('.//td[1]')[0].xpath('string(.)')
        breafinfo[th] = td
    # print(breafinfo)
    teamsBreafInfo[teams[i][0]] = breafinfo
print(teamsBreafInfo)
```

```python
team_list = [0,1,2,3,4,5,9,12,18,19]  #17,34
player_list = [4,9,10,3,3,2,3,7,2,3]
for i in range(len(team_list)):
    html = urlopen(url+teams[team_list[i]][1])
    bsObj = BeautifulSoup(html.read())
    if(type(player_list[i])!=list):
        table = bsObj.find_all('table')[player_list[i]]
        df = pd.read_html(str(table))[0]
    else:
        frames = []
        for j in player_list[i]:
            table = bsObj.find_all('table')[player_list[i]]
            df = pd.read_html(str(table))[0]
            frames.append(df)
    df.to_csv('wiki-player//'+teams[team_list[i]][0]+'.csv')
```

```python
# 加入百度名字，并对wiki名分割
def baidu_name_append(file1,file2):
    df1 = pd.read_csv(file1)
    df2 = pd.read_csv(file2,encoding='GBK')
    # print(df1)
    # print('df2: ', df2)
    for i in range(len(df1)):
        try:
            df1['号码'][i] = int(df1['号码'][i])
            df1['号码'][i] = int(df1['号码'][i])
        except:
            print('error')
    df3 = df1[['号码','球员']]

    # print(type(df3['号码'][0]))
    for i in range(len(df3)):
        flag = type(df3['号码'][i])!=int
        # print(flag)
        if(flag):
            df3.drop(i,axis=0,inplace=True)
    df3['name_baidu'] = [' '] * df3.shape[0]
    df3['en_name'] = [' '] * df3.shape[0]
    df3.reset_index(drop=True, inplace=True)
    for i in range(len(df3)):
        for j in range(len(df2)):
            if(df3['号码'][i]==df2['number'][j]):
                df3['name_baidu'][i] = df2['name'][j]
    zh_name = []
    en_name = []
    for i in np.array(df3['球员'])[:]:
        zh_name.append(i.split('（')[0])
        en_name.append(i.split('（')[-1].split('）')[0])
    for i in range(len(df3)):
        df3['球员'][i] = zh_name[i]
        df3['en_name'][i] = en_name[i]
    return df3
```

```python
def baidu_name_append_valencia(file1,file2):
    df1 = pd.read_csv(file1)
    df2 = pd.read_csv(file2,encoding='GBK')
    for i in range(len(df1)):
        try:
            df1['编号'][i] = int(df1['编号'][i])
        except:
            print('error')
    df3 = df1[['编号','球员名字']]
    for i in range(len(df3)):
        flag = type(df3['编号'][i])!=int
        # print(flag)
        if(flag):
            df3.drop(i,axis=0,inplace=True)
    df3['name_baidu'] = [' '] * df3.shape[0]
    df3['en_name'] = [' '] * df3.shape[0]
    df3.reset_index(drop=True, inplace=True)
    for i in range(len(df3)):
        for j in range(len(df2)):
            if(df3['编号'][i]==df2['number'][j]):
                df3['name_baidu'][i] = df2['name'][j]
    zh_name = []
    en_name = []
    for i in np.array(df3['球员名字'])[:]:
        zh_name.append(i.split('（')[0])
        en_name.append(i.split('（')[-1].split('）')[0])
    for i in range(len(df3)):
        df3['球员名字'][i] = zh_name[i]
        df3['en_name'][i] = en_name[i]
    return df3
```
